```python
import pandas as pd
import numpy as np
from kmodes.kprototypes import KPrototypes
import sklearn
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_pickle('pre_famd_df.pkl')
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df[['Freq','Age']] = df[['Freq','Age']].astype('float32')
```

```python
# name columns with integers so k-prototype can read
df_int_col = df.copy()
df_int_col.columns = range(df.shape[1])
```

```python
df_int_col.head(3)
```

```python
# fit k prototype model
kproto = KPrototypes(n_clusters=8, init='Cao', n_init = 5, verbose=1)
clusters = kproto.fit_predict(df_int_col, categorical = [1,3,4,5,6,7,8,9])
```

```python
# get the distribution of clusters
labels = pd.DataFrame(clusters)
labels[0].value_counts()
```

```python
labels.shape
```

## summarizing the clusters

```python
allvar = pd.read_pickle('allvariables.pkl')
```

```python
allvar.reset_index(inplace=True)
```

```python
# append cluster labels to customer df
allvar['clusterID'] = labels
```

```python
allvar.head()
```

```python
allvar.to_pickle('kproto_allvar.pkl')
```

```python
categorical_data = allvar.select_dtypes(include = object)
categorical_data.columns.to_list()
```

```python
def cluster_summary(cluster_id):
    cluster = allvar[allvar.clusterID==cluster_id]
    
    cluster_summary = cluster[categorical_data.columns].mode().to_dict(orient="records")[0]
    cluster_summary.update(cluster.mean().to_dict())
    return cluster_summary
```

```python
def cluster_comparison(*cluster_ids):
    summaries = []
    for cluster_id in cluster_ids:
        summaries.append(cluster_summary(cluster_id))
    return pd.DataFrame(summaries).set_index("clusterID").T

summary = cluster_comparison(0,1,2,3,4,5,6,7)
```

```python
summary.to_pickle('kproto_summary.pkl')
```
