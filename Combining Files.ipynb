### Combine Files

```python
import pandas as pd
```

```python
#read 2017 csv
df2017 = pd.read_csv("Data 4/2017.csv")  
```

```python
#read 2016 csv
df2016 = pd.read_csv("Data 4/2016.csv")
#read 2015 csv
df2015 = pd.read_csv("Data 4/2015.csv")
#read 2014 csv
df2014 = pd.read_csv("Data 4/2014.csv")
```

```python
#concatenate the frames 
frames = [df2017, df2016, df2015, df2014]
result = pd.concat(frames) 
```

```python
#convert to csv 
result.to_csv('combinedyears.csv')
```
